```python
# M√©trica simple de tiempo promedio de tr√°mite
def tiempo_promedio(dias):
    return round(sum(dias)/max(len(dias),1), 2)
tiempo_promedio([5, 7, 4, 12])
```

## üíº Valor en Sector P√∫blico
- Tiempos de tr√°mite ‚Üì y satisfacci√≥n ciudadana ‚Üë.

```python
# Pr√°ctica: tiempos de tr√°mite antes/despu√©s (toy)
import statistics as st
base = [7, 10, 12, 5, 9]
mejora = [5, 7, 9, 4, 6]
print('Promedio base (d√≠as):', round(st.mean(base),1))
print('Promedio mejorado (d√≠as):', round(st.mean(mejora),1))
print('Reducci√≥n %:', round(100*(st.mean(base)-st.mean(mejora))/st.mean(base),1))
```
