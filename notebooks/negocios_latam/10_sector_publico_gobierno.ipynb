````python
# M√©trica simple de tiempo promedio de tr√°mite
def tiempo_promedio(dias):
    return round(sum(dias)/max(len(dias),1), 2)
tiempo_promedio([5, 7, 4, 12])
````

- Transparencia y datos abiertos ‚Üí confianza y control social.

````python
# Pr√°ctica: tiempos de tr√°mite antes/despu√©s (toy)
import statistics as st
base = [7, 10, 12, 5, 9]
mejora = [5, 7, 9, 4, 6]
print('Promedio base (d√≠as):', round(st.mean(base),1))
print('Promedio mejorado (d√≠as):', round(st.mean(mejora),1))
print('Reducci√≥n %:', round(100*(st.mean(base)-st.mean(mejora))/st.mean(base),1))
````

````python
# Visual: distribuci√≥n de tiempos de tr√°mite
import plotly.express as px
import pandas as pd
vis = pd.DataFrame({'grupo':['base']*len(base)+['mejora']*len(mejora), 'dias':base+mejora})
fig = px.box(vis, x='grupo', y='dias', title='Tiempos de tr√°mite: base vs mejora')
fig.show()
````

## üìã Caso de uso: Transformaci√≥n digital de tr√°mites con interoperabilidad

**Contexto de negocio:** Un organismo p√∫blico tiene tiempo promedio de tr√°mite de 12 d√≠as y 38% de satisfacci√≥n ciudadana. Objetivo: <5 d√≠as y satisfacci√≥n >70%.

**Problema sin ingenier√≠a de datos:**
- Registros en 8 sistemas distintos sin integraci√≥n.
- Ciudadano debe presentar el mismo documento 3 veces en distintas ventanillas.
- Sin visibilidad del estado del tr√°mite.

**Soluci√≥n con ingenier√≠a de datos:**
1. Lakehouse por dominios: padrones, registros civiles, catastro, hacienda con modelo can√≥nico.
2. APIs de interoperabilidad: sistema A consulta dato de sistema B sin duplicar.
3. Portal ciudadano: tracking en tiempo real del estado del tr√°mite.
4. RPA + validaciones autom√°ticas: 70% de tr√°mites sin intervenci√≥n humana.
5. Data contracts: cada dominio publica datasets con SLA y ownership claro.

**Resultado:**
- Tiempo promedio de tr√°mite: 12 d√≠as ‚Üí 4.2 d√≠as.
- Satisfacci√≥n ciudadana: 38% ‚Üí 74%.
- Costos operativos: -35% por automatizaci√≥n.
- Transparencia: auditor√≠a puede rastrear cada consulta y decisi√≥n con linaje completo.
- Casos de fraude detectados: +120% por cruce de datos antes imposible.


````python
# Ejercicio: interoperabilidad entre sistemas de gobierno
import pandas as pd

# Simular registros en 3 sistemas distintos
padron_electoral = pd.DataFrame({
    'dni': ['12345678', '87654321', '11223344'],
    'nombre': ['Juan P√©rez', 'Ana Garc√≠a', 'Carlos L√≥pez'],
    'domicilio': ['Calle A 123', 'Calle B 456', 'Calle C 789'],
})

registro_impuestos = pd.DataFrame({
    'dni': ['12345678', '11223344'],
    'renta_anual': [50000, 80000],
    'categoria': ['Monotributista', 'Aut√≥nomo'],
})

# Integraci√≥n con data contracts (join por clave √∫nica)
integracion = padron_electoral.merge(registro_impuestos, on='dni', how='left')
integracion['estado_fiscal'] = integracion['categoria'].fillna('Sin registro')

print("Vista integrada (ciudadano no necesita presentar datos 2 veces):")
print(integracion[['dni', 'nombre', 'estado_fiscal', 'renta_anual']])
print(f"\nCobertura de integraci√≥n: {integracion['renta_anual'].notna().mean():.1%}")
print("üí° API de interoperabilidad permite que sistema de subsidios consulte renta sin duplicar datos")
````
