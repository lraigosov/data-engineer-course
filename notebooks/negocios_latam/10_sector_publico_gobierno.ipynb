````python
# Métrica simple de tiempo promedio de trámite
def tiempo_promedio(dias):
    return round(sum(dias)/max(len(dias),1), 2)
tiempo_promedio([5, 7, 4, 12])
````

````python
# Práctica: tiempos de trámite antes/después (toy)
import statistics as st
base = [7, 10, 12, 5, 9]
mejora = [5, 7, 9, 4, 6]
print('Promedio base (días):', round(st.mean(base),1))
print('Promedio mejorado (días):', round(st.mean(mejora),1))
print('Reducción %:', round(100*(st.mean(base)-st.mean(mejora))/st.mean(base),1))
````

## 📋 Caso de uso: Transformación digital de trámites con interoperabilidad

**Contexto de negocio:** Un organismo público tiene tiempo promedio de trámite de 12 días y 38% de satisfacción ciudadana. Objetivo: <5 días y satisfacción >70%.

**Problema sin ingeniería de datos:**
- Registros en 8 sistemas distintos sin integración.
- Ciudadano debe presentar el mismo documento 3 veces en distintas ventanillas.
- Sin visibilidad del estado del trámite.

**Solución con ingeniería de datos:**
1. Lakehouse por dominios: padrones, registros civiles, catastro, hacienda con modelo canónico.
2. APIs de interoperabilidad: sistema A consulta dato de sistema B sin duplicar.
3. Portal ciudadano: tracking en tiempo real del estado del trámite.
4. RPA + validaciones automáticas: 70% de trámites sin intervención humana.
5. Data contracts: cada dominio publica datasets con SLA y ownership claro.

**Resultado:**
- Tiempo promedio de trámite: 12 días → 4.2 días.
- Satisfacción ciudadana: 38% → 74%.
- Costos operativos: -35% por automatización.
- Transparencia: auditoría puede rastrear cada consulta y decisión con linaje completo.
- Casos de fraude detectados: +120% por cruce de datos antes imposible.


````python
# Ejercicio: interoperabilidad entre sistemas de gobierno
import pandas as pd

# Simular registros en 3 sistemas distintos
padron_electoral = pd.DataFrame({
    'dni': ['12345678', '87654321', '11223344'],
    'nombre': ['Juan Pérez', 'Ana García', 'Carlos López'],
    'domicilio': ['Calle A 123', 'Calle B 456', 'Calle C 789'],
})

registro_impuestos = pd.DataFrame({
    'dni': ['12345678', '11223344'],
    'renta_anual': [50000, 80000],
    'categoria': ['Monotributista', 'Autónomo'],
})

# Integración con data contracts (join por clave única)
integracion = padron_electoral.merge(registro_impuestos, on='dni', how='left')
integracion['estado_fiscal'] = integracion['categoria'].fillna('Sin registro')

print("Vista integrada (ciudadano no necesita presentar datos 2 veces):")
print(integracion[['dni', 'nombre', 'estado_fiscal', 'renta_anual']])
print(f"\nCobertura de integración: {integracion['renta_anual'].notna().mean():.1%}")
print("💡 API de interoperabilidad permite que sistema de subsidios consulte renta sin duplicar datos")
````
